# Converting to US time

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Nov 19 11:44:02 2021

@author: ashba
"""

import us
import pytz
from datetime import datetime


def getlocaltime(time_zones,dt ):
    try: 
        dt = datetime.strptime(dt, '%H:%M')
        return pytz.timezone(time_zones).fromutc(dt)
    except:
        print (dt)
        
        
        
for_train_data['time_zones'] = for_train_data.apply(lambda row:us.states.lookup(row.state).time_zones[0], axis=1)
for_train_data['localtime'] = for_train_data.apply(  lambda row: getlocaltime(row.time_zones, row.time) , axis=1)
```
