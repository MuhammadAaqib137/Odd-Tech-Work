```python
import pandas as pd
```

```python
new_list=[x for x in data['Bot Transcription'].astype(str) if len(x)<2]
```

```python
new_list
```

```python
new = pd.DataFrame({'Bot Transcription':new_list})
new
```

```python
new_data = pd.merge(data, new, on=["Bot Transcription"], how="outer", indicator=True)

print(new_data)


new_data = new_data.loc[new_data["_merge"] == "left_only"].drop("_merge", axis=1)

print(new_data)

```

```python
new_data
```

```python
new_data['Bot Transcription'].duplicated().value_counts()
```

```python
new_data.drop_duplicates(subset ='Bot Transcription',keep=False,inplace=True)
```

```python
new_data['Bot Transcription'].duplicated().value_counts()
```

```python
new_data
```

```python
new_data=new_data.reset_index(drop=True)
```

```python
new_data
```

# removing data having text after thge beep

```python
junk = []
count = 1
for i in new_data['Bot Transcription'].astype(str):
    if ('after the be' in i)or('press' in i )or('a e e ' in i)or ('e e e' in i) or ('e  a a' in i)or('o o e' in i)or('a a a a' in i)or ('i o e' in i)or('h o e' in i)or('o o o o'in i) or ('a a a a'in i) :
        print('\n '+str(count)+ ' '+i)
        junk.append(i)
        count+=1
```

```python
junk
```

```python
df_rmv_beep = new_data

for i in junk:
    df_rmv_beep = df_rmv_beep[df_rmv_beep['Bot Transcription'] != i]
```

```python
df_rmv_beep
```

```python
df_rmv_beep=df_rmv_beep.reset_index(drop=True)
```

```python
df_rmv_beep
```

```python
junk1 = []
count = 0
for i in df_rmv_beep['Bot Transcription'].astype(str):
    if ('connected 'in i) or ('disconnected 'in i) or ( 'pound'  in i)or('voicemail'in i)or('record your message' in i):
        print('\n '+str(count)+ ' '+i)
        junk1.append(i)
        count+=1
```

```python
noise_remv=df_rmv_beep

for i in junk1:
    noise_remv = noise_remv[noise_remv['Bot Transcription'] != i]
```

```python
noise_remv
```

```python
noise_remv=noise_remv.reset_index(drop=True)
```

```python
noise_remv
```

```python
junk2 = []
count = 0
for i in df_rmv_beep['Bot Transcription'].astype(str):
    if ('avalaible'in i) or ('hospital'in i) or ( 'hang up'  in i)or('mailbox'in i)or('thankyou ' in i)or('a a a' in i)or ('e e a' in i):
        print('\n '+str(count)+ ' '+i)
        junk2.append(i)
        count+=1
```
