```python
import sqlite3
conn = sqlite3.connect('my_database.db')
c = conn.cursor()
```

```python
c.execute('''Drop TABLE table_2018''')
```

```python
c.execute('''CREATE TABLE table_2018 (FullPhoneNumber text,state text,cityname text,persongender  text,personexactage text,personmaritalstatus text,traveler text,ZipCode text,investment text,estimatedincomecode text,timezone text,DONOTCALL text,"PresenceOfCreditCard" text,CreditRating  text )''')
```

```python
import os
list_of_dir = os.listdir(r'Z:\AIDAILER\2018.3')
list_of_iter = iter(list_of_dir)
```

```python
for i in range(len(list_of_dir)):
#     try:
    df1 = pd.read_csv('Z:/AIDAILER/2018.3/'+str(next(list_of_iter)) ,warn_bad_lines=True, error_bad_lines=False, engine="python")
    df1['FullPhoneNumber'] = df1['FullPhoneNumber'].astype(str)
    df1 = df1[df1['FullPhoneNumber'] != 'nan'] 
    df1['ZipCode']=df1['ZipCode'].astype(str)
    df1["DONOTCALL"]=df1["DONOTCALL"].astype(str)
    df1["DONOTCALL"] =df1["DONOTCALL"].map({True: 'Y', False: 'N'})
    df1 = df1[["state","ZipCode","cityname","timezone","FullPhoneNumber","estimatedincomecode","persongender","personexactage","personmaritalstatus","PresenceOfCreditCard","CreditRating","investment","traveler","DONOTCALL"]]
    df1.to_sql('table_2018',conn,if_exists='append',index=False)
    print(str(i+1), str(df1.shape), str(list_of_dir[i]), 'Appended to Database')

#     except Exception as e:
#         print(str(e))
#         pass
```
