# importing libraries

```python
import pandas as pd
import json 
```

# loading Data

```python
data = pd.read_csv('dataset_9.csv',header= None)
```

```python
data
```

# Renaming Columns

```python
data = data.rename({0: 'data.date', 1: 'data.number', 2:'data.application',3:'data.id',4:'data.rawTranscription'}, axis=1)
```

```python
data.drop(5,axis=1,inplace=True)
```

```python
data
```

# changing type of rawTranscription

```python
# changing type of rawTranscription
data['data.rawTranscription'] = data['data.rawTranscription'].apply(lambda x : json.loads(x))
```

```python
data1=data.explode('data.rawTranscription')
data1
```

# audio name

```python
data1['audio_name'] = data1['data.application'] +'-'+ data1['data.id'].astype(str)+'-in.wav'   
```

# chunk name

```python
chunk_name = []
def chunk(x):
    for count,val in enumerate(x):
        chunk_name.append(str(count+1))
```

```python
data['data.rawTranscription'].apply(chunk)
```
