```python
import pandas as pd
```

```python
data = pd.read_csv('Cleaned_CSV.csv')
```

```python
data
```

```python
junk = []
count = 0
for i in data['Bot Transcription'].astype(str):
    if  ('leave a method'in i)or('leave a message'in i)or('leave me your'in i)or ('take your call please' in i )or('after the tone'in i)or ('AT the tone'in i ) or('after recording' in i)or ('after ring' in i):
        print('\n '+str(count)+ ' '+i)
        junk.append(i)
        count+=1
```

```python
data_clean=data
for i in junk:
    data_clean = data_clean[data_clean['Bot Transcription'] != i]
data_clean
```

```python
data_clean= data_clean.reset_index(drop=True)
data_clean
```
